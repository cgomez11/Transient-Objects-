```python
from bs4 import BeautifulSoup
import urllib
import wget
```

```python
def download_images(url):
    request = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(request, "lxml")
    letters = soup.find_all("tr")
    print(len(letters))
    lobbying = {}
    for element in letters:
        lobbying[element.a.get_text()] = {}
        #len(lobbying.keys())
    total_extra_ims = 0
    for x in range(len(lobbying.keys())):
    #for x in range(5):    
        if x > 0: #0 is the header
            s = letters[x].a["onclick"] #where window.open is
            start =s.find('http')
            end = s.find('.html',start)
            new_url = s[start:end+5]
            print('Row URL:',new_url)
            u = urllib.request.urlopen(new_url).read()
            soup_u = BeautifulSoup(u, "lxml")
            ims = soup_u.find_all("img")
            mark = new_url.rfind('/')
            base_url = new_url[0:mark] + '/'
            l = [] #create an empty list 
            for link in ims:
                complete_url = base_url + link.get('src')
                #filename = wget.download(complete_url)
                l.append(complete_url) #l is a list in which each element is the url of an image
            #print(l) #show all the list with urls 
            del l[-1]
            print('***************************')
            #Extra images
            second_url = s[start:end]+'x'+'.html'
            v = urllib.request.urlopen(second_url).read()
            soup_v = BeautifulSoup(v, "lxml")
            ims_extra = soup_v.find_all("img")
            
            if ims_extra == []: #the html code of second_url is empty
                continue 
            #just for the objects where second_url is not empty 
            l_extra = []
            for li in ims_extra:
                complete_url_extra = base_url + li.get('src')[2:end]
                l_extra.append(complete_url_extra)
            #print('Extra Images to download from: \n') 
            #print(l_extra)
            #print(len(l_extra))
    total_extra_ims = total_extra_ims + len(l_extra)        
    print('Total extra ims:', total_extra_ims)
                    
```
